```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from google.colab import files 
uploaded = files.upload()
```

```python
image = cv2.imread('background.jpg', cv2.IMREAD_GRAYSCALE) #그레이스케일로 불러와서

images = [] #빈리스트 생성후 임계값 설정 변경된 각각의 이미지들을 추가 
ret, thres1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY) # 127임계를 넘으면 백,아니면 흑
ret, thres2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV) # 반대
ret, thres3 = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC) #127임계 넘으면 127, 아니면 그대로
ret, thres4 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO) # 127임계 넘으면 그대로 아니면 0
ret, thres5 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)# 위 반대 
images.append(thres1)
images.append(thres2)
images.append(thres3)
images.append(thres4)
images.append(thres5)

for i in images:
  plt.imshow(cv2.cvtColor(i, cv2.COLOR_GRAY2RGB))
  plt.show()
```
