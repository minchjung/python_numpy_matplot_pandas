```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
key = cv2.imread('keyboard.jpg')
back = cv2.imread('background.jpg')
plt.imshow(cv2.cvtColor(key, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(back, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# \cv2.resize로 사이즈 조절 
back2 =cv2.resize(back,None,fx=3.0,fy=3.0,interpolation=cv2.INTER_CUBIC);
key2 = cv2.resize(key,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
```

```python
plt.imshow(cv2.cvtColor(key2, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(back2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#openCV add 연산 
print(back.shape)
print(key.shape)
result = cv2.add(back[0:900,:], key[:,0:950])
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
