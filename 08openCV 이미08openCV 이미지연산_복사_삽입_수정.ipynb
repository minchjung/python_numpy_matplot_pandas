```python
import cv2
# imread <- 이미지를 불러 넘파이 객체로 반환
image = cv2.imread('3.jpg')
# 픽셀 수 및 이미지 크기  
# shape : 넘파이 객체의 행렬 구조 
# size : 넘파이 객체의 길이 
print(image.shape)
print(image.size)
# 이미지 Numpy객체의 특정 픽셀
px = image[100,100]
#B, G, R 순서 
print(px) 
# R만 출력 
print(px[2])
```

```python
import matplotlib.pyplot as plt
# 이미지 슬라이싱 추출 
roi = image[0:500, 0:500]
# 슬라이싱 영역 지정후 추출 부분 삽입 
image[940:1440, 580:1080] = roi

# 넘파이 객체 bgr 모드를 rgb로 컬러 정보 변경 후   
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
# plt 객체로 출력
plt.show()
```

```python
#image[행,열,BGR]
image[:,:,2]=0
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
